```python
print("Variance in age:%f"%np.var(haberman.age))

print("Range in age:[%d,%d]"%(max(haberman.age),min(haberman.age)))
print("Range in no. of lymph nodes:[%d,%d]\n"%(max(haberman.auxilliry_lymph_node),min(haberman.auxilliry_lymph_node)))

print("Standard deviation in age:%f"%np.std(haberman.age))
```

```python
haberman.describe()
```

```python
sns.FacetGrid(haberman, hue = "survival_status", size = 5).map(sns.distplot, "age").add_legend()
plt.title("Histogram of age")
plt.ylabel("Density")
plt.show()
```

```python
sns.FacetGrid(haberman, hue = "survival_status", size = 5).map(sns.distplot,"operation_year").add_legend()
plt.title("Histogram of age")
plt.ylabel("Density")
plt.show()
```

```python
sns.FacetGrid(haberman, hue = "survival_status", size = 5).map(sns.distplot, "auxilliry_lymph_node").add_legend()
plt.title("Histogram of axillary_lymph_node")
plt.ylabel("Density")
plt.show()
```

```python
sns.boxplot(x = "survival_status", y = "age", hue = "survival_status", data = haberman).set_title("Box plot for survival_status and age")
plt.show()
```

```python
sns.boxplot(x = "survival_status", y = "auxilliry_lymph_node", hue = "survival_status", data = haberman).set_title("Box plot for survival_status and axillary_lymph_node")
plt.show()
```
